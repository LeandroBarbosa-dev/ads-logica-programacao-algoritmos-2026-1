### 4. MANIPULAÇÃO AVANÇADAS COM SRINGS


Concatenação

```python
s1 = 'Lógica de programação'
s1 = s1 + ' e Algoritmos'
print(s1)
```

Repetindo strings na concatenação

```python
s1 = 'A' + '-' * 10 + 'B'
print(s1)
```

```python
s1 = 'A' + ' ' * 10 + 'B'
print(s1)
```

Composição por marcador de posição

```python
nota = 8.9
s1 = 'Você tirou %f na disciplina de Algoritmos' % nota
print(s1)
```

Limitando as casas decimais:

```python
nota = 8.5
s1 = 'Você tirou %.2f na disiplina de Algoritmos' % nota
print(s1)
```

Várias variáveis:

```python
nota = 8.5
disciplina = 'Algoritmos'
s1 = 'Você tirou %.2f na disciplina de %s' % (nota, disciplina)
print(s1)
```

Composição moderna

```python
nota = 8.5
disciplina = 'Algoritmos'
s1 = 'Você tirou {} na disciplina de {}'.format(nota, disciplina)
print(s1)
```

Composição com f-string

```python
nota = 8.5
disciplina = 'Algoritmos'
s1 = f'Você tirou {nota} na disciplina de {disciplina}'
print(s1)
```

Fatiamento

```python
s1 = 'Lógica de Programção e Algoritmos'
print(s1[0:6])
```

```python
s1 = 'Lógica de Programção e Algoritmos'
print(s1[24:34])
```

```python
s1 = 'Lógica de Programção e Algoritmos'
print(s1[:6])
```

Tamanho (length)

```python
s1 = 'Lógica de Programação e Algoritmos'
tamanho = len(s1)
print(tamanho)
```
